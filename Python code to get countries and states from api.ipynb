```python
#to make a request on this url 
#to get all the countries
url="https://geodata.solutions/restapi"
#setting the header to json format
headers = {'Content-Type': 'application/json; charset=utf-8'}
#making a post request without any parameters to get the countries
response = requests.request("POST", url, headers=headers)
#getting into json format
r = response.json()
print(r)
```

```python
%%time

#function created to save the data into csv
def store_to_csv(a,response_received):
    #a is the numeric parameter on which we will be looping
    #response_received is the data received from response.json()
    key=str(a)
    #converting integer a to string to give it to the dictionary
    x=response_received[key].get('country_name')
    #fetching the value of country on the basis of the value received
    #url for post request to get the states
    url="https://geodata.solutions/restapi"
    #request.head
    headers = {'Content-Type': 'application/json; charset=utf-8'}
    #sending a request with the parameters based on the the api and x will be the country string
    response=requests.post(url,data={"country":x})
    #changing the response into json again
    r1 = response.json()
    #creating the filename and looping on it to save the data into csv file
    filename='data__states_countries_from_rest.csv'
    for i in range(0,len(r1["details"]['regionalBlocs'])):
        print(response_received[key].get('country_name')+":"+r1['details']['regionalBlocs'][i].get("state_name"))
        fields=[response_received[key].get('country_name'),r1['details']['regionalBlocs'][i].get("state_name")]
        with open(filename, 'a+') as f:
            writer = csv.writer(f)
            writer.writerow(fields)
            f.close()


#calling our function for the number of countries returned
for i in range(0,220):
    store_to_csv(i,r)
```

```python
import pandas as pd
data=pd.read_csv('data__states_countries_from_rest.csv',names=["country","states","unknown"])
columnnames=data.columns
print(columnnames)
data.drop(["unknown"],inplace=True,axis=1)
data.dropna(axis=0,inplace=True)
data.reset_index(inplace=True,drop=True)
# data.to_csv("country and their states list without index.csv",index=False)
# data.to_csv("country and their states list with index.csv",index=False)
#pushing the data from csv to mysql
data["final"]=data["states"]+","+data["country"]
data.to_csv("country_state_merged.csv",index=False)
#data["country"].apply(lambda x : print('{}and{}'.format(x,len(x))))
```
